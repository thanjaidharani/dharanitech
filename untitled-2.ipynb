```python
# 1. Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, classification_report, confusion_matrix, roc_curve)
import joblib

# plotting style
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# 2. Load dataset
df = pd.read_csv(r"C:\Users\Vijay\Documents\Churn.csv")   # update path if required
print("Rows,Cols:", df.shape)
df.head()
```

```python
# 3. Quick audit
df.info()
print("\nMissing values per column:\n", df.isnull().sum())
df.describe().T
```

```python
# 4. Drop irrelevant columns
df = df.drop(columns=['RowNumber', 'CustomerId', 'Surname'], errors='ignore')
```

```python
# 5. Encode categorical features
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df = pd.get_dummies(df, columns=['Geography'], drop_first=True)  # creates Geography_Germany, Geography_Spain if present
df.head()
```

```python
# 6. Features and target
X = df.drop('Exited', axis=1)
y = df['Exited']
print("Class distribution:\n", y.value_counts(normalize=True))
```

```python
# 7. Scaling
num_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])

# Save scaler
joblib.dump(scaler, "scaler.joblib")
```

```python
# 8. Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y)
print(X_train.shape, X_test.shape)
```

```python
# 9. Baseline Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)

# Predictions
y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]

# Metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix visualization
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# ROC curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], '--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```

```python
# 10. Feature importance
feat_imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print(feat_imp.head(15))
feat_imp.head(15).plot(kind='bar', figsize=(10,5))
plt.title("Top feature importances (Random Forest)")
plt.show()
```

```python
# 11a. Try RandomForest with class_weight
rf_balanced = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42, n_jobs=-1)
rf_balanced.fit(X_train, y_train)
y_pred_bal = rf_balanced.predict(X_test)
y_proba_bal = rf_balanced.predict_proba(X_test)[:,1]
print("Recall (balanced RF):", recall_score(y_test, y_pred_bal))
print("ROC-AUC (balanced RF):", roc_auc_score(y_test, y_proba_bal))

# 11b. Threshold tuning (example: lower threshold to raise recall)
threshold = 0.35
y_pred_thresh = (y_proba >= threshold).astype(int)
print("Metrics at threshold", threshold)
print("Recall:", recall_score(y_test, y_pred_thresh))
print("Precision:", precision_score(y_test, y_pred_thresh))
print("F1:", f1_score(y_test, y_pred_thresh))
```

```python
# 12. GridSearchCV (example small grid)
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 8, 12],
    'min_samples_split': [2, 5]
}

gs = GridSearchCV(RandomForestClassifier(random_state=42, n_jobs=-1),
                  param_grid, cv=3, scoring='roc_auc', verbose=1)
gs.fit(X_train, y_train)
print("Best params:", gs.best_params_)
best_rf = gs.best_estimator_

# Evaluate tuned
y_proba_best = best_rf.predict_proba(X_test)[:,1]
print("ROC-AUC (tuned):", roc_auc_score(y_test, y_proba_best))
print("Accuracy (tuned):", accuracy_score(y_test, best_rf.predict(X_test)))
```

```python
# 13. Logistic Regression
log = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
log.fit(X_train, y_train)
y_proba_log = log.predict_proba(X_test)[:,1]
print("ROC-AUC (Logistic):", roc_auc_score(y_test, y_proba_log))
print(classification_report(y_test, (y_proba_log >= 0.5).astype(int)))
```

```python
# 14. Save model(s)
joblib.dump(rf, "rf_churn_baseline.joblib")
joblib.dump(rf_balanced, "rf_churn_balanced.joblib")
joblib.dump(best_rf, "rf_churn_tuned.joblib")   # if GridSearch was run
joblib.dump(log, "logistic_churn.joblib")
# scaler already saved earlier as scaler.joblib
```
